Data preprocessing

Step 1: Missing values

```python
def clean_dataframe(df, drop_threshold=0.4, fill_threshold=0.1):
    """
    Cleans the dataframe by:
    1. Dropping columns with more than 40% missing values.
    2. Filling columns with less than 10% missing values with the mode (by default) or 
       median for specified columns.
    
    Parameters:
    df (pd.DataFrame): The input dataframe to clean.
    drop_threshold (float): The percentage threshold for dropping columns (default is 40%).
    fill_threshold (float): The percentage threshold for filling columns (default is 10%).
    
    Returns:
    pd.DataFrame: The cleaned dataframe.
    """
    # List of columns to fill with median instead of mode
    median_fill_columns = [
        'Main_Carriageway_Width', 
        'Total_Road_Width', 
        'Year_of_Birt', 
        'Mobile_Obstacle', 
        'Obstacle_Hit'
    ]
    
    # Calculate the percentage of missing values for each column
    missing_percent = df.isnull().mean()
    
    # Drop columns where more than 40% of the values are missing
    df_cleaned = df.drop(columns=missing_percent[missing_percent > drop_threshold].index)
    
    # Iterate through columns to fill missing values based on the name or data type
    for column in df_cleaned.columns:
        missing_fraction = df_cleaned[column].isnull().mean()
        
        if missing_fraction > 0 and missing_fraction < fill_threshold:
            if column in median_fill_columns:
                # Replace missing values with the median for specific columns
                median_value = df_cleaned[column].median()
                df_cleaned[column] = df_cleaned[column].fillna(median_value)
            else:
                # Replace missing values with the mode for all other columns
                mode_value = df_cleaned[column].mode()[0]
                df_cleaned[column] = df_cleaned[column].fillna(mode_value)
    
    return df_cleaned
```

```python
characteristics_df_cl = clean_dataframe(characteristics_df, drop_threshold=0.4, fill_threshold=0.1)
holidays_df_cl = clean_dataframe(holidays_df, drop_threshold=0.4, fill_threshold=0.1)
places_df_cl = clean_dataframe(places_df, drop_threshold=0.4, fill_threshold=0.1)
users_df_cl = clean_dataframe(users_df, drop_threshold=0.4, fill_threshold=0.1)
vehicles_df_cl = clean_dataframe(vehicles_df, drop_threshold=0.4, fill_threshold=0.1)
```

```python
#Address Column: Change NA to 'Unknown'
characteristics_df_cl.Address = characteristics_df_cl.Address.fillna('Unknown')
```

Step 4: Removing redundant/irrelevant columns

```python
# Address (not categorical, lot of unique values) and Commune_Code (covered by Department_Code, too many unique values)
characteristics_df_cl.drop(['Address', 'Commune_Code'], axis = 1, inplace = True)
# Road number (lot of unique values)
places_df_cl.drop(['Road_Number'], axis = 1, inplace = True)
# Vehicle ID
users_df_cl.drop(['Vehicle_ID'], axis = 1, inplace = True)
vehicles_df_cl.drop(['Vehicle_ID'], axis = 1, inplace = True)
# Collision_Type appears both in characteristics_df and vehicles_df
vehicles_df_cl.drop(['Collision_Type'], axis = 1, inplace = True)
```

Step 5: Changing data types

```python
characteristics_df_cl = characteristics_df_cl.astype({
    'Weather_Conditions': 'int', 
    'Collision_Type': 'int',
    'Department_Code': 'int'
})

places_df_cl = places_df_cl.astype({
    'Road_Category': 'int', 
    'Circulation_Type': 'int',
    'Number_of_Lanes': 'int',
    'Road_Profile': 'int',
    'Road_Curvature': 'int',
    'Surface_Condition': 'int',
    'Infrastructure_Type': 'int',
    'Accident_Situation': 'int',
    'Environmental_Factors': 'int',
    'Traffic_Lane_Type' : 'int',
})

users_df_cl = users_df_cl.astype({
    'Position_In_Vehicle': 'int', 
    'Journey_Type': 'int',
    'Safety_Equipment': 'int',
    'Pedestrian_Location': 'int',
    'Pedestrian_Activity': 'int',
    'Pedestrian_Condition': 'int',
    'Year_of_Birth' : 'int'
})

vehicles_df_cl = vehicles_df_cl.astype({
    'Travel_Direction': 'int',
    'Vehicle_Manoeuvre': 'int',
    'Obstacle_Hit' : 'int',
    'Mobile_Obstacle' : 'int'
})
```

Data merging

```python
###Merging characteristics_df with holidays_df

# Convert Date column to Datetime
holidays_df_cl['Date'] = pd.to_datetime(holidays_df_cl['Date'])

# Extract month and day from Date column
holidays_df_cl['Month'] = holidays_df_cl['Date'].dt.month
holidays_df_cl['Day'] = holidays_df_cl['Date'].dt.day
holidays_df_cl.drop('Date', axis = 1, inplace = True)

holidays_df_cl
```

```python
# Merge the two dataframes on 'Month' and 'Day' columns
merged_char_hol = pd.merge(characteristics_df_cl, holidays_df_cl, on=['Month', 'Day'], how='left')
```

```python
# Converting holiday column into binary format
merged_char_hol['Holiday'] = merged_char_hol['Holiday_Name'].notna().astype(int)
merged_char_hol.drop('Holiday_Name', axis = 1, inplace = True)
merged_char_hol.head()
```

```python
# Merging all dataframes on accident ID
merged_df = merged_char_hol.merge(users_df_cl, on='Accident_ID', how='left')\
                            .merge(vehicles_df_cl, on='Accident_ID', how='left')\
                            .merge(places_df_cl, on='Accident_ID', how='left')
```
